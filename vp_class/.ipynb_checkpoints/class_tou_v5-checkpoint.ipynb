This code creates a time-dependent OU process class with the following parameterization:

$dX_{t} = [\frac{2at + b}{-2(at^{2} + bt + c)}(X_{t} - \alpha((t+1)^{\beta}-1)) + \alpha\beta(t+1)^{\beta-2}] dt + dW_{t}$

```python
import torch
import numpy as np
from torch.autograd import Variable
from torch.distributions import Normal


class tOU:
    def __init__(self, t0, t1, parameters, sde_sigma, init_dist, timegrid=False, dN=500):
        """
            t0, t1 = torch.tensor(t1)
            alpha, beta, m0, r, sigma = torch.tensor.shape (D, 1)
        """
        assert all(sigma > 0), "variance must be positive"
        
        self.t0 = t0
        self.t1 = t1
        
        # set parameters
        self.alpha = parameters['alpha']
        self.beta = parameters['beta']
        self.a = parameters['a']
        self.b = parameters['b']
        self.c = parameters['c']
        
        # set hyper-parameter
        self.sigma = sde_sigma
        
        D = self.alpha.shape[0]
        if timegrid == True:
            self.pts = torch.linspace(t0, t1, dN).repeat(D, 1)
        else:
            self.pts = torch.sort(torch.cat([(t1 - t0) * torch.rand(D, dN-2) + t0, torch.tensor([self.t0, self.t1]).repeat(D,1)], axis=1), axis=1)[0]
            
        self.trj, self.dt = self.path(init_dist)
        
            
    def path(self, init_dist):
        """
            init_dist: distribution to draw an initial state
            init_dist = Normal(loc=E[X_{t_{i-1}}|lambda_{i-1}], scale = std[X_{t_{i-1}}|lambda_{i-1}])
        """
        
        # Get parameters
        alpha = self.alpha
        beta = self.beta
        a = self.a
        b = self.b
        c = self.c
        
        sigma = self.sigma
        
        t = self.pts
        
        
        def mean(x, t0, alpha, beta, m0, a, b, c):
            """
                t0: we always start our process from t = 0
            """
            prevtime = t0[:, 0].reshape(-1, 1)
            curtime = t0[:, 1].reshape(-1, 1)
            
            e_alpha = torch.sqrt((a * curtime ** 2 + b * curtime + c) / (a * prevtime ** 2 + b * prevtime + c))                      
            b_t = alpha * ( (prevtime + 1) ** beta  - 1) + m0 - (alpha  * ((curtime + 1) ** beta - 1) + m0 ) * e_alpha
            
            return (x.reshape(-1, 1) - b_t) / e_alpha
        
        def variance(t, a, b, c, sigma):
            prevtime = t[:, :-1]
            curtime = t[:, 1:]
            
            numerator = 2*a*(curtime ** 3 - prevtime ** 3) + 3*b*(curtime ** 2 - prevtime ** 2) + 6 * c * (curtime - prevtime)
            denominator = 6 * (a * curtime ** 2 + b * curtime + c)
            return numerator * sigma ** 2 / denominator 
        
        def std(t, a, b, c, sigma):
            return torch.sqrt(variance(t, a, b, c, sigma))

        
        normal = Normal(loc=0., scale=1.)
        x = normal.sample(t.size())

        if type(init_dist) == torch.distributions.normal.Normal:
            # sample initial state from approximate posterior distribution
            m0 = init_dist.sample([x.shape[0], 1])
            x[:, 0] = m0.flatten()
        else:
            m0 = init_dist
            x[:, 0] = m0.flatten()
            
        t0 = t - t[:, 0].reshape(-1, 1)

        dt = torch.diff(t)

        scale = std(t, a, b, c, sigma)

        x[:, 1:] = x[:, 1:] * scale
        for i in range(1, x.shape[1]):
            x[:, i] += mean(x[:, i-1], t0[:, i-1:i+1], alpha, beta, m0, a, b, c).flatten()
        
        return x, dt
```

```python
dist = Normal(loc=0, scale=1.)
t0 = torch.tensor(1.)
t1 = torch.tensor(2.)
sigma = torch.tensor([1.]).reshape(-1, 1)
```

```python
parameters = {'alpha':torch.tensor([-3.]).reshape(-1, 1),\
             'beta':torch.tensor([1.]).reshape(-1, 1), \
             'a': torch.tensor([-0.001]).reshape(-1, 1), \
             'b': torch.tensor([-0.01]).reshape(-1, 1), \
             'c': torch.tensor([-0.01]).reshape(-1, 1)}
```

```python
init_pts = torch.tensor([1.0]).reshape(-1, 1)
```

```python
tou = tOU(t0, t1, parameters, sigma, init_dist=init_pts)
```

```python
import matplotlib.pylab as plt

plt.plot(tou.pts.T, tou.trj.T)
```

```python
def tou_gradient(t, x, params, sde_sigma, init_dist):
    assert torch.is_tensor(sde_sigma), "sde_sigma must be a (D*1) tensor"
    assert torch.is_tensor(init_dist) or type(init_dist) == torch.distributions.normal.Normal,\
    "init_dist must either be tensor or torch.distributions"
    
    """
        Calculates log likelihood of a path
        Note that there are three parameters, alpha, beta, r
    """
    
    def mean(x, t, alpha, beta, m0, a, b, c):
        """
            t0: we always start our process from t = 0
        """
        t0 = t - t[:, 0].reshape(-1, 1)
        prevtime = t0[:, :-1]
        curtime = t0[:, 1:]
        ratio = (a * curtime ** 2 + b * curtime + c) / (a * prevtime ** 2 + b * prevtime + c)
        
        max_ratio = torch.maximum(ratio, torch.tensor([0.01]))
        
        e_alpha = torch.sqrt(max_ratio)                      
        b_t = alpha * ( (prevtime + 1) ** beta  - 1) + m0 - (alpha  * ((curtime + 1) ** beta - 1) + m0 ) * e_alpha

        return (x - b_t) / e_alpha

    def variance(t, a, b, c, sigma):
        t0 = t - t[:, 0].reshape(-1, 1)
        prevtime = t0[:, :-1]
        curtime = t0[:, 1:]

        numerator = 2*a*(curtime ** 3 - prevtime ** 3) + 3*b*(curtime ** 2 - prevtime ** 2) + 6 * c * (curtime - prevtime)
        denominator = 6 * (a * curtime ** 2 + b * curtime + c)

        return numerator * sigma ** 2 / denominator 

    def std(t, a, b, c, sigma):
        return torch.sqrt(variance(t, a, b, c, sigma))
    
    param_stack = torch.stack([params['alpha'], params['beta'], params['a'], params['b'], params['c']])
    
    params_ = Variable(param_stack, requires_grad=True)
    alpha, beta, a, b, c = params_
    sigma = sde_sigma

    m0 = x[:, 0].reshape(-1, 1)
    mu = mean(x[:, :-1], t, alpha, beta, m0, a, b, c)
    var = std(t, a, b, c, sigma) + 1e-7 # To prevent the underflow (some of the value becomes 0 due to lack of precision
    LL = torch.nansum(Normal(loc=mu, scale=var).log_prob(x[:, 1:]), axis=1)

    #Â At initialization (in case of random initialization)
    if type(init_dist) == torch.distributions.normal.Normal:
        LL += torch.nansum(init_dist.log_prob(x[:,0]))
    
    LL.backward(torch.tensor([1.]).repeat(x.shape[0]))

    return {'alpha':params_.grad[0].clone().detach(), 'beta':params_.grad[1].clone().detach(), 
            'a':params_.grad[2].clone().detach(), 'b':params_.grad[3].clone().detach(), \
            'c':params_.grad[4].clone().detach(), 'LL':LL.clone().detach().data}

```

```python
def Grad_Descent( current_pars, current_grad, params, learning_rate = 0.01):
    """
        Perform vanila gradient descent step
        returns dictionary such that dict["name_of_parameter"] = value after one gradient step
    """
    assert type(current_pars) == type(current_grad) == dict, "Current_grad must be a dict"
    
    new_pars = current_pars.copy()
    for key in current_pars.keys():
        if key in params:
            new_pars[key] = current_pars[key] + learning_rate * current_grad[key]

    return new_pars
```

```python
def MLE(pts, trj, sigma, dist, init_pars, params, IT, learning_rate):
    current_pars = init_pars
    pars_hist = {key:[] for key in current_pars.keys()}
    ll_hist = []
    
    for i in range(IT):
        for key in current_pars.keys(): 
            pars_hist[key].append(current_pars[key].item())
        current_grad = tou_gradient(pts, trj, current_pars, sigma, dist)
        
        ll_hist.append(current_grad['LL'].item())
        
        current_pars = Grad_Descent( current_pars, current_grad, params, learning_rate)
        if i % 50 == 0:
            print("LogLikelihood: {}".format(current_grad['LL']))
    
    return current_pars, pars_hist, ll_hist
```

```python
param_mle, param_hist, ll_hist = MLE(tou.pts, tou.trj, sigma, dist, init_pars, paras_names, 2000, learning_rate=0.03)
```

```python
def get_PostMoment(t0, tT, alpha, beta, m0, vN):
    """
        Returns mean and variance of the approximate posterior process
        Here, we 
        Note: mean and variance have tractable expressions
    """
    # Get time-grid between s0 and sK
    
    t = torch.linspace(0, (tT-t0).item(), vN+2)[1:]
    
    def B(s, beta):
        return (s+1) ** beta
        
    def m(s):
        return alpha * (B(s, beta) - 1) + m0
    
    
    mean = m(t)
    

#     def var(s):
#         return sigma ** 2 / (2 * r) * (1 - torch.exp(-2 * r * s))

#     var = var(t)
        
    return t, mean
```

```python
pts, m_t = get_PostMoment(t0, t1, param_mle['alpha'], param_mle['beta'], init_pts, vN=200 )
pts, true  = get_PostMoment(t0, t1, parameters['alpha'], parameters['beta'], init_pts, vN=200 )
```

```python
plt.plot(pts, m_t.flatten(), label='mle mean')
plt.plot(pts, true.flatten(), label='true mean')
plt.legend()
```
